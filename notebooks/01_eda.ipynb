```python
import pandas as pd
import matplotlib.pyplot as plt

# ==============================
# CARGA DE DATOS
# ==============================

df = pd.read_csv("../data/ventas_videojuegos.csv", sep=';', decimal=',')

print("Shape del dataset:", df.shape)
print(df.head())

# ==============================
# LIMPIEZA BÁSICA
# ==============================

df = df.dropna()

# ==============================
# ANÁLISIS 1: VENTAS POR AÑO
# ==============================

ventas_por_anio = df.groupby("Año")["Ventas Global"].sum()

plt.figure()
ventas_por_anio.plot()
plt.title("Ventas Globales por Año")
plt.xlabel("Año")
plt.ylabel("Ventas Globales")
plt.show()

# ==============================
# ANÁLISIS 2: VENTAS POR GÉNERO
# ==============================

ventas_por_genero = df.groupby("Genero")["Ventas Global"].sum().sort_values(ascending=False)

plt.figure()
ventas_por_genero.plot(kind='bar')
plt.title("Ventas Globales por Género")
plt.xlabel("Género")
plt.ylabel("Ventas")
plt.show()

# ==============================
# ANÁLISIS 3: VENTAS POR PLATAFORMA
# ==============================

ventas_por_plataforma = df.groupby("Plataforma")["Ventas Global"].sum().sort_values(ascending=False).head(10)

plt.figure()
ventas_por_plataforma.plot(kind='bar')
plt.title("Top 10 Plataformas por Ventas")
plt.xlabel("Plataforma")
plt.ylabel("Ventas")
plt.show()
```
